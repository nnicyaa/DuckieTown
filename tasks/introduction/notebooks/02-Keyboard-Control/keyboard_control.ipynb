## How It Works

`get_motor_speeds(keys_pressed)` converts key state to `(left_speed, right_speed)`.

| Key | Left wheel | Right wheel | Effect |
|-----|-----------|------------|--------|
| ↑ | +0.5 | +0.5 | Forward |
| ↓ | -0.5 | -0.5 | Backward |
| ← | -0.3 | +0.3 | Turn left |
| → | +0.3 | -0.3 | Turn right |

Keys combine — pressing ↑ and ← at the same time curves forward-left.


```python
from tasks.introduction.packages.manual_drive import get_motor_speeds
# Try different key combinations
test_cases = [
    ('none',  {'up': False, 'down': False, 'left': False, 'right': False}),
    ('↑',     {'up': True,  'down': False, 'left': False, 'right': False}),
    ('↓',     {'up': False, 'down': True,  'left': False, 'right': False}),
    ('←',     {'up': False, 'down': False, 'left': True,  'right': False}),
    ('→',     {'up': False, 'down': False, 'left': False, 'right': True}),
    ('↑+←',   {'up': True,  'down': False, 'left': True,  'right': False}),
    ('↑+→',   {'up': True,  'down': False, 'left': False, 'right': True}),
]

print(f"{'Keys':<8} {'Left':>8} {'Right':>8}")
print('-' * 26)
for label, keys in test_cases:
    left, right = get_motor_speeds(keys)
    print(f"{label:<8} {left:>+8.2f} {right:>+8.2f}")
```
